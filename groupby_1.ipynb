```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat_1.csv')
```

```python
data.head()
```

```python
data_1 = data.fillna(value = 0)
```

```python
data_1.head()
```

```python
group_mean = data_1.groupby('group', as_index=False).agg({'session_value': 'mean'})
```

```python
group_mean.columns = ['group', 'mean_session_value']
```
